```python
# 1) dla pobranych z klawiatury dwóch liczb zwróci ich iloczyn.
# ENG Solution for two numbers taken will return their product.

a = input("Podaj 1 liczbę - ")

b = input("Podaj 2 liczbę - ")

wynik = int(a)*int(b)

print('Wynik to:', wynik)

# 2) sprawdzi, czy podana liczba jest parzysta.
# ENG will check if the given number is even.

n=raw_input('podaj liczbe')
def jest_parzysta(n):
    x = 0
    if str(n) % 2 == str(x):
        a = 'jest parzysta'
    else:
        a = 'jest nieparzysta'
         
    return a
print ("jest parzysta:",(n))

# 3) obliczy sumę kwadratów kolejnych liczb naturalnych mniejszych od zadanej wartości.
# ENG calculates the sum of squares of consecutive natural numbers smaller than a given value. 
def suma(n):
    s = 0
    for i in range(n):
        s = s** + i
    return s
print(suma(6))

# 4) sprawdzi, czy podana liczba jest pierwsza.
# ENG will check if the given number is prime. 
n = int(input("Podaj liczbę naturalną: "))
def czy_pierwsza(n):
    if n == 2:
        return "Liczba  jest parzysta"
    if n % 2 == 0 or n <= 1:
        print("Liczba nie jest parzysta")
print(czy_pierwsza(n))

# 5) znajdzie dzielniki podanej liczby.
# ENG finds the divisors of the given number.
print("Program do obliczania dzielników liczby.")

a = int(input("Podaj liczbe do sprawdzenia: "))

def dzielniki(n):
    for i in range(1, int(n/2)+1):
	    if n % i == 0:
	        print("Dzielniki liczby "+ str(n) +" to: "+str(i))
    print("Dzielniki liczby: "+ str(n) +" to: "+ str(n))

print(dzielniki(a))

# 6) zwróci rozkład zadanej liczby na dzielniki pierwsze.
# ENG returns the decomposition of the given number into prime divisors.
n = int(input("podaj liczbę do rozłożenia na czynniki: "))

def rozloz(n):
    czynniki = []
    k = 2
    while n != 1:
        while n % k == 0:
            n //= k
            czynniki.append(k)
        k += 1
    
    return czynniki

print(rozloz(n))
# zad 7 znajdzie najmniejszą wspólna wielokrotność podanych dwóch liczb.
# ENG finds the smallest common multiple of the given two numbers.
a = int(input("Podaj 1 liczbę:"))

b = int(input("Podaj 2 liczbę:"))


def nwd(a, b):
    
	if b > 0:
        
		return nwd(b, a%b)
    
		return a
    

def nww(a, b): 
    
	return a*b//nwd(a, b)
print(nww(a,b))

# zad 8 przedstawi zadaną liczbę w postaci binarnej.
# ENG will represent the given number in binary form.
n = int(input("Podaj liczbę:"))
print(bin(n)[2:])

# zad 9 dla zadanych wartości współczynników trójmianu kwadratowego znajdzie jego miejsca zerowe.
# ENG for given values of the coefficients of the quadratic triangle will find its zero places. 
print("Dla równania ax^2+bx-c=0,")
a = int(input("Podaj współczynnik a funkcji kwadratowej:"))
b = int(input("Podaj współczynnik b funkcji kwadratowej:"))
c = int(input("Podaj współczynnik c funkcji kwadratowej:"))


delta = (b*b)-(4*a*c)
if delta > 0:
    x1 = (-b-delta*(1/2))/(2*a)
    x2 = (-b+delta*(1/2))/(2*a)
    print ('x1 = '), x1, ("x2= "), x2
elif delta == 0:
    x0 = int(-b/(2*a))
    print ("x0 = ",x0)
else:
    print ('brak rozwiązań')

# zad 10 dla danego n obliczy n! . Podaj różne rozwiązania tego problemu.
# ENG for a given n will calculate n! Give different solutions to this problem.
def silnia(n):
    wynik = 1
    for i in range(1, n+1):
        wynik *= i
    return wynik

# zad 10 cz.2 
# ENG part.2
import math

def silnia(n):
    return math.factorial(n)

# zad 11 oblicza n-tą liczbę Fibonacciego (n duże).
# ENG Calculates the nth Fibonacci number (n large).
print("Podaj n-tą liczbę ciągu Fibbonaciego:")

n = int(input())

def fibo(n):
    if (n == 0):
        return 0
    if (n == 1):
        return 1
    return fibo(n-1)+fibo(n-2)
print(fibo(n))

# zad 12 dla dowolnej ilości argumentów zwróci ich skumulowane sumy częściowe.
# ENG for any number of arguments will return their cumulative subtotals.
def cumulative_sum(*args):
    cum_sum = []
    total_sum = 0
    for arg in args:
        total_sum += arg
        cum_sum.append(total_sum)
    return cum_sum

user_input = input("Wprowadź liczby oddzielone przecinkami: ")
numbers = [int(x.strip()) for x in user_input.split(",")]
result = cumulative_sum(*numbers)
print(result)

# zad 13 dla dowolnej ilości argumentów liczbowych zwróci ich średnią arytmetyczną.
# ENG for any number of numeric arguments will return their arithmetic mean. 
print("Podaj z ilu liczb, obliczyć średnią")
n = int(input())
suma = 0

for i in range(1, n+1):
       print("Podaj liczbę " + str(i))
       a = float(input())
       suma += a
    if n <=0:
    print("Podałeś liczb mniejszą od 0")
    
print(i)
    
print("średnia ocen to " + str(suma/n))

# zad 14 dla zadanego łancucha znaków zwróci łańcuch ze znakami w odwrotnej kolejności.
# ENG for the given string will return a string with characters in reverse order.
print("Podaj ciąg znaków do odwrócenia:")

txt = input()

newtext = txt [::-1]
print(newtext)

# zad 15 dla danego łańcucha znaków zwróci informację ile jest w nim dużych, a ile małych liter.
# ENG for a given string will return information on how many uppercase and lowercase letters are in it.
print("Podaj zdanie do policzenia małych i dużych liter:")
n = str(input())

lower = 0
upper = 0
a = n.strip()
for i in a:
      if(i.islower()):
            lower +=1
      elif(i.isupper()):
            upper += 1
      else:
            continue

print("Liczba małych liter:",lower)
print("Liczba dużych liter:",upper)

# zad 16 dla danej tablicy zwróci tablicę z jej unikalnymi elementami.
# ENG for a given array will return an array with its unique elements. 
print("Podana tablica: ")
thislist = ["pomarańczowy","niebieski","fioletowy"]
print(thislist)

print("Jaki znak chcesz dodać do listy: ")
n = str(input())

thislist.append(n)

print(thislist)

# zad 17 dla zadanego łańcucha znaków zwróci słownik ze statystykami na temat częstości występowania poszczególnych znaków.
# ENG for the given string will return a dictionary with statistics on the frequency of each character.

def char_frequency(string):
    freq = {}
    for char in string:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    return freq

string = "Hello, World!"
freq = char_frequency(string)
print(freq)

# zad 18 dla zadanego stringa zwróci liste słów, na które można go podzielić.
# ENG for the given string will return a list of words into which it can be divided.
print("Podaj zdanie do rozdzielenia:")
n = str(input())
words = n.split(",")
print(words)

# zad 19 dla zadanego string zwróci jego postać w notacji wielbłądziej, np. dla ”ala ma pythona” otrzymamy ’alaMaPythona’.
# ENG for the given string will return its form in camel notation.
print("Podaj zdanie do zmiany: ")

n = str(input())

a = n.title()

print(a)

# zad 20 dla stringa w postaci wielbłądziej zwróci jego postać ”normalną”, np. dla ’alaMaPythona’ otrzymamy ”ala ma pythona”.
# ENG for a string in camel form will return its "normal" form.
import re

def sentence_case(string):
    if string != '':
        result = re.sub('([A-Z])', r' \1', string)
        return result[:1].upper() + result[1:].lower()
    return


text = "PrzykładowyJedenTekst"
print(sentence_case(text))

# zad 21 zgodnie z metodą babilońską obliczy przybliżenie pierwiastka kwadratowego danej liczby.
# ENG according to the Babylonian method will calculate the approximation of the square root of a given number.
print("Podaj liczbę początkową:")
p=int(input())

print("Podaj ilość iterracji:")
a = int(input())

def babilon(p,a):

    for i in range(a):
        p = (p + a / p) / 2
        a = a / p

    return p
print("Przybliżona wartość: ")
print(babilon(p,a))

# zad 22 przybliża wartość π (metoda Monte-Carlo).
# ENG approximates the value of π (Monte-Carlo method).
import random
def monte_carlo_pi(num_points):

    inside_circle = 0


    for i in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x ** 2 + y ** 2 <= 1:
            inside_circle += 1

    pi = 4 * inside_circle / num_points

    return pi

print(monte_carlo_pi(1000000))

# zad 23 dla list a = [1, 2, 3] i b = [4, 5, 6] zwróci wynik c = [(1, 4),(2, 5),(3, 6)].
# ENG for lists a = [1, 2, 3] and b = [4, 5, 6] will return the result c = [(1, 4),(2, 5),(3, 6)].
print("Lista nr1: ")
first = ["1", "2", "3"]
print(first)

print("Lista nr2: ")
second = ["4", "5","6"]
print(second)
a = first[0],second[0]
b = first[1],second[1]
c = first[2],second[2]
third = [a,b,c]

print("Nowa lista")
print(third)

# zad 24 dla zakresu liczb range(100) zwróci listę wartości, które są podzielne przez 3 i niepodzielne przez 5.
# ENG for a range of numbers range(100) will return a list of values that are divisible by 3 and indivisible by 5.
print("Zakres od 0 do 100")
print("Liczby podzielne przez 3 i nie podzielne przez 5:")

for i in range(0,100):
    if i % 5 != 0 and i % 3 ==0:
        print(i)
    else:
        continue

print(i)
```
